```python
# Step 1: Import necessary libraries
import os
import pandas as pd
import numpy as np
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
from IPython.display import display, Image
from PIL import Image as PILImage


# Step 2: Define function to extract features from tokenized tweets
def extract_features(tokenized_tweets):
    model = Word2Vec(tokenized_tweets, vector_size=100, window=5, min_count=1, workers=4)
    word_vectors = model.wv
    features = []
    for tokens in tokenized_tweets:
        vectors = [word_vectors[word] for word in tokens if word in word_vectors]
        if vectors:
            feature_vector = np.mean(vectors, axis=0)
        else:
            feature_vector = np.zeros(100)
        features.append(feature_vector)
    return np.array(features)

# Step 4: Define function to cluster tweets using K-Means
def kmeans_cluster(features, num_clusters):
    km = KMeans(n_clusters=num_clusters, random_state=0)
    km.fit(features)
    clusters = km.labels_.tolist()
    return km, clusters

# Step 5: Determine optimal number of clusters using elbow method
def plot_elbow(features, save_path):
    Sum_of_squared_distances = []
    K = range(1, 10)
    for k in K:
        km = KMeans(n_clusters=k, random_state=0)
        km.fit(features)
        Sum_of_squared_distances.append(km.inertia_)
    plt.plot(K, Sum_of_squared_distances, 'bx-')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Sum of squared distances')
    plt.title('Elbow Method For Optimal Number of Clusters')
    plt.xticks(K)
    
    # Find optimal k 
    diff = np.diff(Sum_of_squared_distances)
    diff_r = diff[1:] / diff[:-1]
    k_opt = K[np.argmin(diff_r) + 1]
    
    # Add vertical line and dot at optimal k
    plt.axvline(x=k_opt, linestyle='--', color='red', label='Optimal k')
    
    # Add legend
    plt.legend()
    
    # Save the plot
    plt.savefig(save_path)
    plt.show()
    plt.close()
    
    return k_opt

# Step 3: Get a list of all the CSV files in the source folder
csv_folder_path = r'D:\Research\Python\Data\WIP\Spike Data'
csv_files = [file for file in os.listdir(csv_folder_path) if file.endswith('.csv')]

# Create a DataFrame to store results
results_df = pd.DataFrame(columns=['Dataset', 'Optimal_Clusters'])

# Step 6: Loop through each CSV file and perform K-means clustering
results_list = []  # Initialize a list to store results
elbow_images = []  # Initialize a list to store elbow plot images
clustering_images = []  # Initialize a list to store clustering representation images
for csv_file in csv_files:
    file_path = os.path.join(csv_folder_path, csv_file)
    df = pd.read_csv(file_path)
    # Drop rows with missing values in the 'Text_Lemmatized' column
    df.dropna(subset=['Text_Lemmatized'], inplace=True)
    tokenized_tweets = df['Text_Lemmatized'].apply(lambda x: x.split())
    features = extract_features(tokenized_tweets)
    
    # Create a folder to save the data for each dataset
    save_folder_path = os.path.join(r'D:\Research\Python\Data\WIP\K-Means Cluster Data', csv_file[:-4])
    os.makedirs(save_folder_path, exist_ok=True)
    
    # Save the elbow plot image
    elbow_plot_save_path = os.path.join(save_folder_path, 'elbow_plot.jpeg')
    optimal_clusters = plot_elbow(features, elbow_plot_save_path)
    
    # Cluster tweets using optimal number of clusters
    km, clusters = kmeans_cluster(features, optimal_clusters)
    
    # Save the clustering representation plot
    clustering_plot_save_path = os.path.join(save_folder_path, 'clustering_representation.jpeg')
    plt.figure(figsize=(10, 8))
    plt.scatter(features[:, 0], features[:, 1], c=clusters, cmap='rainbow', s=20)
    plt.title(f'Clustering Representation - {csv_file[:-4]}')
    plt.xlabel('Feature Dimension 1')
    plt.ylabel('Feature Dimension 2')
    plt.savefig(clustering_plot_save_path)
    plt.close()
    
    # Save the dataset name and the optimal number of clusters in the results list
    results_list.append({'Dataset': csv_file, 'Optimal_Clusters': optimal_clusters})
    
    # Append the images to the respective lists for combining later
    elbow_images.append(PILImage.open(elbow_plot_save_path))
    clustering_images.append(PILImage.open(clustering_plot_save_path))

# Combine elbow plots of all datasets into one 4x3 grid image
elbow_combined_image = PILImage.new('RGB', (1600, 1200))
for i, img in enumerate(elbow_images):
    elbow_combined_image.paste(img.resize((400, 300)), ((i // 3) * 400, (i % 3) * 300))
elbow_combined_image.save(r'D:\Research\Python\Data\WIP\K-Means Cluster Data\elbow_plots_combined.jpeg')

# Combine clustering representation plots of all datasets into one 4x3 grid image
clustering_combined_image = PILImage.new('RGB', (3200, 2400))
for i, img in enumerate(clustering_images):
    clustering_combined_image.paste(img.resize((800, 600)), ((i // 3) * 800, (i % 3) * 600))
clustering_combined_image.save(r'D:\Research\Python\Data\WIP\K-Means Cluster Data\clustering_plots_combined.jpeg')

# Create a DataFrame from the results list
results_df = pd.DataFrame(results_list)

# Save the results DataFrame as a CSV
results_csv_path = r'D:\Research\Python\Data\WIP\K-Means Cluster Data\results.csv'
results_df.to_csv(results_csv_path, index=False)
```
